# Example of Importing Aggregation
An example of how to import the aggregation code so that there is no need to copy and paste. Note that you have to be working within the country_A folder for this import to work.

```python
# importing the code as a module
from aggregation import aggregation
```

```python
import pickle
import pandas as pd
```

```python
# Load data
with open('indiv.pickle', 'rb') as file:
    indiv = pickle.load(file)
with open('hhold.pickle', 'rb') as file:
    hhold = pickle.load(file)
```

```python
X = indiv.set_index(['id', 'iid']).drop(labels=['poor'], axis=1)
Y = pd.DataFrame(hhold.set_index(['id'])['poor'])
```

```python
# Example of choosing to review only specific aggregate functions
sample2 = aggregation.Aggregate(X, Y=Y, groupby='id')
sample2.fit(num_agg_funcs=['median', 'mean'], cat_agg_funcs=['percent', 'any'])
sample2.col_to_func
```
